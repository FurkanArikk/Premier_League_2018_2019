# Soccer Data
This dataset contains data of every game from the 2018-2019 season in the English Premier League.

## Data Dictionary
| Column    | Explanation                                   |
| --------- | --------------------------------------------- |
| Div       | Division the game was played in               |                            
| Date      | The date the game was played                  |
| HomeTeam  | The home team                                 |
| AwayTeam  | The away team                                 |
| FTHG      | Full time home goals                          |
| FTAG      | Full time away goals                          |
| FTR       | Full time result                              |
| HTHG      | Half time home goals                          |
| HTAG      | Half time away goals                          |
| HTR       | Half time result                              |
| Referee   | The referee of the game                       |
| HS        | Number of shots taken by home team            |
| AS        | Number of shots taken by away team            |
| HST       | Number of shots taken by home team on target  |
| AST       | Number of shots taken by away team on target  |
| HF        | Number of fouls made by home team             |
| AF        | Number of fouls made by away team             |
| HC        | Number of corners taken by home team          |
| AC        | Number of corners taken by away team          |
| HY        | Number of yellow cards received by home team  |
| AY        | Number of yellow cards received by away team  |
| HR        | Number of red cards received by home team     |
| AR        | Number of red cards received by away team     |

[Source](https://data.world/chas/2018-2019-premier-league-matches) of dataset.

# 1.1 Outcome Analysis:

## 1.1.1 Find the teams with the most wins in their home and away matches

```python
# Find the team whit the most and the lowest win at home 
home_team_most = df[df['FTR']== 'H']['HomeTeam'].value_counts().idxmax()
home_team_min = df[df['FTR'] == 'H']['HomeTeam'].value_counts().idxmin()

# Find the team whit the most and the lowest win at away 
away_team_most = df[df['FTR'] == 'A']['AwayTeam'].value_counts().idxmax()
away_team_min = df[df['FTR'] == 'A']['AwayTeam'].value_counts().idxmin()

# Display the results
print("The Team with most home win: ", home_team_most)
print("The Team with lowest home win: ", home_team_min)
print("The Team with most away win: ", away_team_most)
print("The Team with lowest away win: ", away_team_min)
```

## 1.1.2 Calculate the percentage of home and away wins

```python
# Calculate the win rate for each team
home_win_rate = df[df['FTR'] == 'H']['HomeTeam'].value_counts() / df['HomeTeam'].value_counts()
away_win_rate = df[df['FTR'] == 'A']['AwayTeam'].value_counts() / df['AwayTeam'].value_counts()

# Find the team with the highest win rate at home and away
best_home_team = home_win_rate.idxmax()
best_away_team = away_win_rate.idxmax()

# Calculate the win rate percentage at home for the best home team and away team
best_home_win_rate = home_win_rate.max() * 100
best_away_win_rate = away_win_rate.max() * 100

# Display the results
best_home_team, best_home_win_rate, best_away_team, best_away_win_rate
```

# 1.4 Create the League Table

```python
# Calculate the number of matches played by each team
matches_played = df['HomeTeam'].append(df['AwayTeam']).value_counts().reset_index()
matches_played.columns = ['Team', 'Matches Played']

# Calculate the number of home wins for each team
home_wins = df[df['FTR'] == 'H']['HomeTeam'].value_counts().reset_index()
home_wins.columns = ['Team', 'Home Wins']

# Calculate the number of away wins for each team
away_wins = df[df['FTR'] == 'A']['AwayTeam'].value_counts().reset_index()
away_wins.columns = ['Team', 'Away Wins']

# Calculate the number of draws for each team
draws = df[df['FTR'] == 'D']['HomeTeam'].append(df[df['FTR'] == 'D']['AwayTeam']).value_counts().reset_index()
draws.columns = ['Team', 'Draws']

# Calculate the number of losses for each team
losses = matches_played.merge(home_wins, on='Team', how='left').merge(away_wins, on='Team', how='left').merge(draws, on='Team', how='left')
losses = losses.fillna(0)
losses['Losses'] = losses['Matches Played'] - losses['Home Wins'] - losses['Away Wins']

# Calculate the total points for each team
points = (home_wins['Home Wins'] + away_wins['Away Wins']) * 3 + draws['Draws']
points_df = pd.DataFrame({'Team': home_wins['Team'], 'Points': points})

# Merge all the team statistics
team_stats = matches_played.merge(home_wins, on='Team', how='left').merge(away_wins, on='Team', how='left').merge(draws, on='Team', how='left').merge(losses[['Team', 'Losses']], on='Team', how='left').merge(points_df, on='Team', how='left')

# Sort the teams by points in descending order
team_stats = team_stats.sort_values(by='Points', ascending=False).reset_index(drop=True)

# Display the results
team_stats
```
